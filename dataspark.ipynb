```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("Sales.csv")
```

```python
df1
```

```python
df1=df1.drop(columns=["Delivery Date"])
```

```python
df2=pd.read_csv("Stores.csv")
```

```python
df2["Square Meters"]=df2["Square Meters"].fillna(0.0)
```

```python
df2
```

```python
df3=pd.read_csv("Customers.csv", encoding="latin1")
```

```python
df3
```

```python
df4=pd.read_csv("Exchange_Rates.csv")
```

```python
df4
```

```python
df5=pd.read_csv("Products.csv")
```

```python
df5
```

```python
merged_df = pd.merge(df1, df2, on='StoreKey', how='left')
```

```python
merged_df= pd.merge(merged_df, df3, on='CustomerKey', how='left')
```

```python
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])
df4['Date'] = pd.to_datetime(df4['Date'])
```

```python
merged_df = pd.merge(merged_df, df4, left_on=['Order Date', 'Currency Code'], right_on=['Date', 'Currency'], how='left')
```

```python
merged_df = merged_df.drop(columns=['Date', 'Currency'])
```

```python
merged_df = pd.merge(merged_df, df5, on='ProductKey', how='left')
```

```python
df=merged_df
```

```python
df
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["state_code"]=df["state_code"].fillna("NA")
```

```python
df.to_csv("dataspark.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Distribution of order quantities
plt.figure(figsize=(10, 6))
sns.histplot(df['quantity'], kde=True)
plt.title('Distribution of Order Quantities')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Orders by Country
plt.figure(figsize=(12, 6))
sns.countplot(y='country_x', data=df, order=df['country_x'].value_counts().index)
plt.title('Number of Orders by Country')
plt.xlabel('Number of Orders')
plt.ylabel('Country')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Number of orders per year
orders_per_year = df['order_date'].dt.year.value_counts().sort_index()
plt.figure(figsize=(10, 6))
orders_per_year.plot(kind='bar')
plt.title('Number of Orders per Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.show()
```
