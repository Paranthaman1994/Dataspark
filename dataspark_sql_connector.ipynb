```python
df["open_date"] = pd.to_datetime(df["open_date"], errors='coerce')
df = df.where(pd.notnull(df), None)  # Convert NaN to None
```

```python
create_table_query= """CREATE TABLE orders (
    order_number INT,
    line_item INT,
    order_date DATE,
    customerkey INT,
    storekey INT,
    productkey INT,
    quantity INT,
    currency_code VARCHAR(10),
    country_x VARCHAR(50),
    state_x VARCHAR(50),
    square_meters FLOAT,
    open_date DATE,
    gender VARCHAR(10),
    name VARCHAR(255),
    city VARCHAR(100),
    state_code VARCHAR(50),
    state_y VARCHAR(50),
    zip_code VARCHAR(20),
    country_y VARCHAR(50),
    continent VARCHAR(50),
    birthday DATE,
    exchange FLOAT,
    product_name VARCHAR(255),
    brand VARCHAR(100),
    color VARCHAR(50),
    unit_cost_usd VARCHAR(50),
    unit_price_usd VARCHAR(50),
    subcategorykey INT,
    subcategory VARCHAR(255),
    categorykey INT,
    category VARCHAR(255)
);
"""
```

```python
cursor.execute(create_table_query)
conn.commit()
```

```python
insert_sales_exchange_values = '''
    INSERT INTO orders (
        order_number, 
        line_item, 
        order_date, 
        customerkey, 
        storekey, 
        productkey, 
        quantity, 
        currency_code, 
        country_x, 
        state_x, 
        square_meters, 
        open_date, 
        gender, 
        name, 
        city, 
        state_code, 
        state_y, 
        zip_code, 
        country_y, 
        continent, 
        birthday, 
        exchange, 
        product_name, 
        brand, 
        color, 
        unit_cost_usd, 
        unit_price_usd, 
        subcategorykey, 
        subcategory, 
        categorykey, 
        category
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s,%s)
'''

# Insert data into the sales_exchange_table
for index, row in df.iterrows():
    cursor.execute(insert_sales_exchange_values, (
        row['order_number'],
        row['line_item'],
        row['order_date'],
        row['customerkey'],
        row['storekey'],
        row['productkey'],
        row['quantity'],
        row['currency_code'],
        row['country_x'],
        row['state_x'],
        row['square_meters'],
        row['open_date'],
        row['gender'],
        row['name'],
        row['city'],
        row['state_code'],
        row['state_y'],
        row['zip_code'],
        row['country_y'],
        row['continent'],
        row['birthday'],
        row['exchange'],
        row['product_name'],
        row['brand'],
        row['color'],
        row['unit_cost_usd'],
        row['unit_price_usd'],
        row['subcategorykey'],
        row['subcategory'],
        row['categorykey'],
        row['category']
    ))
conn.commit()
```
